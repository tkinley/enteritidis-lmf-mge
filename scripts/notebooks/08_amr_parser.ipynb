```python
import pandas as pd
import os
import glob

# Path to the AMRFinder results directory
# Expected structure: result/<sample_id>/<sample_id>.amrfinder.tsv
base_path = 'result' 

# Glob pattern targets all TSV files one directory level below 'result'
search_pattern = os.path.join(base_path, "*", "*.tsv")

files = glob.glob(search_pattern)

print(f"Found {len(files)} AMRFinder files.")

if len(files) == 0:
    print("No files found. Verify that the working directory contains the 'result' folder.")
    print(f"Current working directory: {os.getcwd()}")
    print(f"Search pattern used: {search_pattern}")

all_amr = []

for f in files:
    try:
        # Derive the sample identifier from the parent directory name
        # e.g., result/GCA_002038285.1/GCA_002038285.1.amrfinder.tsv -> GCA_002038285.1
        sample_id = os.path.basename(os.path.dirname(f))
        
        df = pd.read_csv(f, sep='\t')
        
        df.insert(0, 'Sample', sample_id)
        
        # Retain only the columns relevant to downstream analysis
        cols_to_keep = ['Sample', 'Gene symbol', 'Sequence name', 'Element type', 'Class', 'Subclass', 'Method']
        available_cols = [c for c in cols_to_keep if c in df.columns]
        df = df[available_cols]
        
        all_amr.append(df)
        
    except Exception as e:
        print(f"Error reading {f}: {e}")

# Concatenate per-sample DataFrames and export to CSV
if all_amr:
    master_amr_df = pd.concat(all_amr, ignore_index=True)
    master_amr_df.to_csv('master_amr_results.csv', index=False)
    print("Saved 'master_amr_results.csv'")
    
else:
    print("No data to save.")
```
