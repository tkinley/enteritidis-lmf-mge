```python
import pandas as pd
import numpy as np

# 1) Filter metadata and assign cohort labels
metadata = metadata[metadata['Assembly'] != 'GCA_037003325.1'].copy()
metadata['Cohort'] = metadata['is_LMF'].map({1: 'LMF', 0: 'Control'})

# 2) Normalize mobility labels in MOB-suite results
mob = mob_results.copy()
mob['predicted_mobility'] = mob['predicted_mobility'].astype(str).str.strip().str.lower()

# Harmonize variant spellings to a single canonical label
mob['predicted_mobility'] = mob['predicted_mobility'].replace({
    'non-mobilizable': 'non_mobilizable',
    'non mobilizable': 'non_mobilizable',
    'nonmobilizable': 'non_mobilizable',
})

# 3) Count plasmids per genome
plasmid_counts = mob.groupby('Sample').size().reset_index(name='plasmid_count')

# 4) Derive per-genome binary mobility flags
mobility_sets = mob.groupby('Sample')['predicted_mobility'].apply(lambda s: set(s)).reset_index(name='mob_set')
mobility_sets['has_conjugative'] = mobility_sets['mob_set'].apply(lambda s: int('conjugative' in s))
mobility_sets['has_mobilizable'] = mobility_sets['mob_set'].apply(lambda s: int('mobilizable' in s))
mobility_sets['has_non_mobilizable'] = mobility_sets['mob_set'].apply(lambda s: int('non_mobilizable' in s))
mobility_sets = mobility_sets.drop(columns=['mob_set'])

# 5) Build the genome-level summary DataFrame used by Figure 7
genome_df = (
    metadata[['Assembly', 'SNP cluster', 'Cohort']]
    .merge(plasmid_counts, left_on='Assembly', right_on='Sample', how='left')
    .merge(mobility_sets, left_on='Assembly', right_on='Sample', how='left', suffixes=('', '_mob'))
)

# Drop merge helper columns if present
for col in ['Sample', 'Sample_mob']:
    if col in genome_df.columns:
        genome_df = genome_df.drop(columns=[col])

# Fill zeros for genomes with no MOB-suite rows
fill_cols = ['plasmid_count', 'has_conjugative', 'has_mobilizable', 'has_non_mobilizable']
genome_df[fill_cols] = genome_df[fill_cols].fillna(0).astype(int)

genome_df.head()
```

```python
import pandas as pd
import numpy as np

# ----------------------------
# 1) Define the final QC genome set (64 genomes)
# ----------------------------
qc_genomes = (
    metadata.loc[metadata["Assembly"] != "GCA_037003325.1", "Assembly"]
    .astype(str)
    .unique()
)

# ----------------------------
# 2) Binary genome-level presence matrix for MOB-suite Primary Cluster IDs
#    pc_wide[g, pc] = 1 if genome g has at least one plasmid with Primary Cluster ID pc
# ----------------------------
mob_sub = mob_results.loc[mob_results["Sample"].astype(str).isin(qc_genomes), ["Sample", "primary_cluster_id"]].copy()
mob_sub["Sample"] = mob_sub["Sample"].astype(str)

pc_wide = (
    mob_sub.assign(present=1)
    .pivot_table(index="Sample", columns="primary_cluster_id", values="present",
                 aggfunc="max", fill_value=0)
    .astype(int)
)

# ----------------------------
# 3) Binary genome-level presence matrix for PlasmidFinder replicon families
#    rep_wide[g, rep] = 1 if genome g has replicon family rep
# ----------------------------
rep = pf_wide.copy()
rep["Sample"] = rep["Sample"].astype(str)

# If there are duplicate Sample rows, collapse by max (treat any hit as present)
rep = rep.groupby("Sample", as_index=False).max(numeric_only=True)

rep = rep.loc[rep["Sample"].isin(qc_genomes)].set_index("Sample")
rep = rep.apply(pd.to_numeric, errors="coerce").fillna(0)
rep_wide = (rep > 0).astype(int)

# ----------------------------
# 4) Align genomes present in both matrices
# ----------------------------
common = pc_wide.index.intersection(rep_wide.index)
pc = pc_wide.loc[common].copy()
rep = rep_wide.loc[common].copy()

# ----------------------------
# 5) Compute conditional probabilities:
#    P(replicon present | Primary Cluster ID present)
# ----------------------------
rows = []
for pc_id in pc.columns:
    mask = pc[pc_id] == 1
    denom = int(mask.sum())
    if denom == 0:
        continue

    # conditional probability for each replicon family
    probs = rep.loc[mask].mean(axis=0)

    for rep_name, p in probs.items():
        n_both = int(((rep[rep_name] == 1) & mask).sum())
        rows.append({
            "primary_cluster_id": pc_id,
            "replicon_family": rep_name,
            "n_genomes_with_primary_cluster": denom,
            "n_genomes_with_both": n_both,
            "P(replicon_present | primary_cluster_present)": float(p),
        })

cond_df = pd.DataFrame(rows)

# Optional: filter out tiny denominators (keeps estimates stable)
cond_df_stable = cond_df[cond_df["n_genomes_with_primary_cluster"] >= 3].copy()

# Show top concordant replicons per Primary Cluster ID
(cond_df_stable
 .sort_values(["primary_cluster_id", "P(replicon_present | primary_cluster_present)"],
              ascending=[True, False])
 .head(30)
)
```
