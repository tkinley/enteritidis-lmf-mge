```python
import pandas as pd
import os
import glob

# Set the path to your main mob_suite results folder
# If the notebook is in the same folder as 'mob_suite', use './mob_suite'
base_dir = './mob_suite'

# Create an empty list to store data from each sample
all_results = []

# Walk through each subfolder (e.g., GCA_002038285.1)
# We look for files ending in mobtyper_results.txt
search_pattern = os.path.join(base_dir, "**/mobtyper_results.txt")
files = glob.glob(search_pattern, recursive=True)

print(f"Found {len(files)} MOBtyper result files.")

for f in files:
    # Extract the Assembly/Sample ID from the folder name
    sample_id = os.path.basename(os.path.dirname(f))
    
    # Read the tab-separated mobtyper file
    try:
        df_temp = pd.read_csv(f, sep='\t')
        # Add the Sample ID as a column for identification
        df_temp.insert(0, 'Sample', sample_id)
        all_results.append(df_temp)
    except Exception as e:
        print(f"Could not read {f}: {e}")

# Combine all individual dataframes into one master table
master_mob_df = pd.concat(all_results, ignore_index=True)

# Save to CSV for your statistics
master_mob_df.to_csv('master_mobtyper_results.csv', index=False)

# Display a preview
print("Master Table Preview:")
display(master_mob_df.head())
```
