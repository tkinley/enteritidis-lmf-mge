```python
import pandas as pd
import os
import glob

# Path to the MOB-suite output directory
base_dir = './mob_suite'

# Container for per-sample result DataFrames
all_results = []

# Search recursively for MOBtyper result files across all sample subdirectories
search_pattern = os.path.join(base_dir, "**/mobtyper_results.txt")
files = glob.glob(search_pattern, recursive=True)

print(f"Found {len(files)} MOBtyper result files.")

for f in files:
    # Derive the sample identifier from the parent directory name
    sample_id = os.path.basename(os.path.dirname(f))
    
    try:
        df_temp = pd.read_csv(f, sep='\t')
        # Prepend the sample identifier as the first column for downstream identification
        df_temp.insert(0, 'Sample', sample_id)
        all_results.append(df_temp)
    except Exception as e:
        print(f"Could not read {f}: {e}")

# Concatenate all per-sample DataFrames into a single master table
master_mob_df = pd.concat(all_results, ignore_index=True)

# Export the consolidated results to CSV
master_mob_df.to_csv('master_mobtyper_results.csv', index=False)

print("Master Table Preview:")
display(master_mob_df.head())
```
